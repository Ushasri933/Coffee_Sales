```
CS['Day']=pd.to_datetime(CS['date'])
daily_sales=CS.groupby('Day')['money'].sum().reset_index()
daily_sales
```

```
CS['Month']=CS['date'].dt.to_period('M')
monthly_sales=CS.groupby('Month')['money'].sum().reset_index()
monthly_sales
```

```
CS['Year']=CS['date'].dt.year
yearly_sales=CS.groupby('Year')['money'].sum().reset_index()
yearly_sales
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(25,6))
plt.plot(daily_sales['Day'],daily_sales['money'],marker='o',color='g')
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Money')
plt.grid(True)
plt.show()
```

```
plt.plot(monthly_sales['Month'].astype(str),monthly_sales['money'],marker='D',linestyle='dashed',color='g')
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Money')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

```
plt.plot(yearly_sales['Year'],yearly_sales['money'],marker='o',color='g')
plt.title('Yearly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Money')
plt.show()
```
